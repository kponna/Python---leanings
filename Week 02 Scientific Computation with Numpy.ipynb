# 2. Creation of Arrays 

### Types of Array and its Creation
There are two main general mechanism of creating arrays.
1. Converting python sequences to numpy arrays:
Numpy arrays can be created from various Python sequences, such as lists and tuples, where python lists are defined using squara brackets eg:[1,2,3,4] and parentheses for tuples eg:(1,2,3,4).
These sequences can be used to created numpy ndarrays with different dimensions.

Any array object created using numpy is referred to as an ndarray, where the 'nd' stands for 'n-dimensional'. This terminology is used to differentiate numpy arrays from standard python sequences such as lists and tuples.

2. Array creation function:
Numpy provides several functions for creating arrays with specific values or patterns

1) np.zeros: Creates an array filled with zeros

2) np.ones: Creates an array filled with ones

3) np.empty: Creates an array without initializing its values (i.e, the values are unintialized and may contain random data)

4) np.arange: Creates an array with evenly spaced values withing a specified interval

5) np.linspace: Creates an array with a specified num of evenly spaced values within a specified interval

6) np.logspace: Creates an array with a specified num of logarithmically spaced values within a specified interval

7) np.full(): Creates an array of specified shape and fills it with a specified value.

8) np.eye(): Creates an identity matrix, which is a square matrix with ones on the diagonal and zeros elsewhere

# 3. Basic operations on Arrays

```python
## Arithematic operations

import numpy as np
a = np.array([1,2,3,4,5])
b = np.array([4,2,8,9,6])
# Addition
print(a+b)

# Subtraction
print(a-b)

# Multiplication
print(a*b)

# Division
print(a/b)

# Modulo - gives reminder
print (a%b)

# Exponentiation
print(a**b) # a raised to power b
```

```python
# matrix multiplication - dimensional array
c = np.array([[1,2],[3,4]]) #2d arrays
d = np.array([[7,2],[9,4]])
print(np.dot(c,d))
```

```python
# trignometric functions
trig = np.array([0,15,30,60,90])
# degree to radians
trig1 = np.deg2rad(trig)
#converting the degree values to radians
print('Sin:', np.sin(trig1))
print('Cos:', np.cos(trig1))
```

```python
# Logarithmic functions
print(np.log(a))
print(np.log10(a))
```

```python
# Bitwise operations - or, and,.....
bit1 = np.array([1,2,3,4,5], dtype = np.uint8) #uint8 is unsignedinteger
bit2 = np.array([6,7,8,9,5], dtype = np.uint8)
print('Bitwise and :', np.bitwise_and(bit1,bit2))
print('Bitwise or :', np.bitwise_or(bit1,bit2))
```

```python
## COmparision operators - It does element wise check
a = np.array([1,2,3,4,5])
b = np.array([4,2,8,9,6])
print(a==b)
print(a>b)
print(a<b)
print(a!=b)
print(a>=b)
print(a<=b)
```

```python
## Aggregation functions

# Sum
print(np.sum(a))
# Mean
print(np.mean(a))
# Median
print(np.median(a))
# Variance
print(np.var(a))
# Standard deviation
print(np.std(a))
# Maximum
print(np.amax(a))
# Minimum
print(np.amin(a))
```

```python
## Sorting Operations 
print(np.sort(a)) #sorts in ascending order
print(np.sort(a)[::-1]) #using slicing sorts in descending order
```

# 6. Broadcasting

Broadcasting is a feature in the numpy library in python that allows for arithmetic operations between arrays of different shapes. 

Broadcasting solves the problem of compatibility in shape between arrays of different dimensions during arithmetic operations. 

The smaller array is broadcasted across the larger array so that they have compatible shapes. 

Broadcasting provides a means of vectorizing arry operations so that looping occurs in C instead of python.

Broadcasting is nothing but stretching

```python
arr1 = np.array([2,4,5,6,7,8])
arr2 = np.array([[2,3],[1,5]])
arr = arr1 + arr2
print(arr)
#we get the error as they both are of different dimensions
```

```python
# Multiplying a scalar to a bigger array-- simple broadcasting
```

```python
a = np.array([2,3,4,5,6,7])
b = 2.00
c = a*b
c #see fig 1
```

### General Rules of Broadcasting

1. Axes must be aligned.

If the arrays have different numbers of dimensions, the shape of the smaller array will be padded with ones on the left.

```python
a = np.array([1,4])
b = np.array([[4],[6],[7]])
a.shape
b.shape
c = a+b
c
```

2. The size in each dimension must either be equal or one of them must be one.

If the size in a particular dimension does not match and neither is one, then a valueerror will be raised.

```python
a = np.array([[2,3,4], [5,6,7]])  # here the elements in each array should be equal
b = np.array([4,6,7])
c = a+b
c
```

3. Broadcasting occurs along the first dimension

If an array has size 1 in a particular dimension and the other array has a size greater than 1, then the smaller array's shape is broadcast to match the larger array.

```python
a = np.array([[2,3,4], [5,6,7]]) #2*3 matrix no of elements in each list tells the no of cols
b = np.array([4,16,7]) #1*3 matrix
c = a+b
c 
# here the shape of b is broadcasted to the shape of a to perform the addition resulting in the larger array
```

## Limitations

1. Broadcasting only works for arrays with a lower dimension to an array with a higher dimension, not vice versa.
2. Broadcasting can only add or multiply a scalar to an array, it cannot perform other operations such as division or subtraction.
3. Broadcasting is a performance-critical operation and can lead to slow computation if performed incorrectly, such as when broadcasting arrays with a large number of elements.

# 8. I/O Handling with Numpy

I/O handling refers to the process of inputting and outputing data to and from a computer system.
This includes reading data from a variety of sources, such as files and databases, and writing data to different types of storage, such as hard drives or cloud storage. 

I/O handling is crucial aspect of computer programming as it allows programs to interact with the outside world and manipulate data. 

Numpy provides several functions for I/O handling:

1. numpy.loadtxt(): This function is used to load data from a text file or a CSV file into a numpy array.


2. numpy.genfromtxt(): This function is used to load data from a text file or a csv file into a numpy array, and can handle more complex data structures such as missing values, variable num of colmns, etc.

3. numpy.savetxt(): This function is used to save data from a numpy array to a text file or a csv file.

4. numpy.save(): This function is used to save data from a numpy array to a binary file. The data can be loaded later using the numpy.load() function.

5. numpy.load(): This function is used to load data from a binary file created using numpy.save().


```python
import numpy as np
```

```python
# saving data to text file
data = np.array([[1,2,3],[4,5,6]])
np.savetxt('datadoc.txt',data,delimiter=',') #it creates a text file if datadoc file didnt existed 

loaded_data = np.loadtxt('datadoc.txt',delimiter=',') #loading the data from the text file
#saving the data into a binary file
np.save('data.npy',data)
#loading this binary file
loaded_data = np.load('data.npy')
```

# 9. Masking of Arrays

Masking of arrays in Numpy involves creating a Boolean mask (an array of the same shape as the original array,with each element being either True or false) to filter out or manipulate specific elements based on a condition , by using the mask to index the original array.

True elements represent the elements of the original array that we want to keep. while the false element represents the elements that we want to remove.

```python
import numpy as np
```

```python
arr = np.array([2,3,4,5,6])
```

```python
# creating a mask
mask = arr > 3
result = arr[mask]
print(result) #['False','False','True','True','True']
```

```python
arr1 = np.array([-2,-1,0,1,2])
mask = arr1 <0
arr1[mask] = 0 #replacing the elements <0 with 0
arr1
#mask = ['True','True','False','False','False']
```

# 10. Structured Arrays

Structured arrays are ndarrays whose datatype is a composition of simpler datatypes organized as a sequence of named fields. Structured arrays can be a powerful tool for organizing , processing , and analyzing structured data, and are an important tool for many data analysis and scientific computing applications.


Features of structured features of structured arrays:


1. Data types of fields: The data type of each field in a structured array can be specified using the standard Numpy data types such as int, float, complex, bool, str, etc. In addition, you can also use custom data types such as datetime64, timedelta64 etc


2. Structured arrays and nested structures: You can create nested structures within structured arrays, by specifying the datatype of a field as another structured array. This allows you to create more complex data structures such as trees, graphs etc


3. Flexible field names: the names of the field in a structured array do not have to be unique. This allows you to create arrays with duplicate fields, which can be used for in certain situations. 


4. Sub arrays: you can extract a subarray from a structured array, just like you would extract a subarray from regular Numpy array. This can be useful when you only need to work with a portion of the data in a structured array.


5. compatibility with other libraries: structured arrays in Numpy can be easily converted to and from the other data formats such as CSV, JSON, HDF5,etc. This makes it easy to integrate Numpy structured arrays with other libraries and tools.


```python
import numpy as np
```

```python
#Define the datatype of each field
dt = np.dtype([('name','S10'), ('age','i4'), ('height','f8')])  
#S10 -string with length of 10, i4 is int of 4 bytes, f8 is float of 8 bytes
#name, age, height are the fields

#create the structured array 
people = np.array([('John', 23, 2.44), ('Jane', 45.5, 22)], dtype = dt) #we can pass n no of records using the array
# we can pass heterogenous data in the arrays
people
```

```python
#Access the first record
person = people[0]

#Access the name field of the first record
name = person['name'].decode('utf-8') # Assuming 'name' is encoded as bytes This will decode the bytes using UTF-8 encoding and print the result as a regular string without the 'b' prefix.
print(name)
#Access the age field of the first record
age = person['age']
print(age)
#Access the height field of the first record
height = person['height']
print(height)
```

```python
# Calculate the average height of all people
avg_height = people['height'].mean()
print(avg_height)

# Find all people who are older than 30
older_than_30 = people[people['age']> 30]
print(older_than_30) 
```

```python
# other feature
# Define a custom datatype for dates
date_dtype = np.dtype([('year','i4'),('month','i2'), ('day', 'i2')])

#Define the datatype for each field
dt = np.dtype([('name','S10'),('age','i4'),('birthdate',date_dtype), ('height','f8')])

#Create the structured array
people = np.array([('John', 32, (1990, 12, 25), 1.56), ('Jane', 28 , (1993, 2, 15), 1.64)], dtype = dt) #birth date is passed as tuple
people
```

```python
# Accessing the nested lists - same as dictionary
# Access the birthdate of the first person
birthdate = people[0]['birthdate']
print(birthdate)

# Access the year of birth of the first person
year_of_birth = birthdate['year']
print(year_of_birth)
```

```python
#sorting structured arrays
sorted_people = np.sort(people, order = 'age') # sorted in asc order of age
print(sorted_people)
```

```python
# modifying fields
#increase the age of all people by 2
people['age'] += 1
print(people['age'])

#change the height of the first person
people[0]['height'] = 1.89
print(people[0]['height'])
```
